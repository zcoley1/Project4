```python
#Word Frequency Dictionary
f = open("./bee-movie-2.txt", "r")
bee_movie = f.read()
words = bee_movie.lower().split()
print(words)
```

```python
wordFreqs = {}

for word in words: 
    if word in wordFreqs:
        wordFreqs[word] += 1
    else:
        wordFreqs[word] = 1
print(wordFreqs)
```

```python
def most_common_words(wordFreqs, k): 
    sorted_words = sorted(wordFreqs.items(), key=lambda x: (-x[1], x[0]))
    common_words = [word for word, freq in sorted_words[:k]]
    return common_words
k = 4
top_words = most_common_words(wordFreqs, k)
print(top_words)
```

```python
top_three_most_frequent = ['barry', 'the', 'a']
top_five_most_frequent = ['barry', 'the', 'a', 'and', 'to']
top_ten_most_frequent = ['barry', 'the', 'a', 'and', 'to', 'you', 'vanessa', 'i', 'it', 'is']

# Test 1
if top_three_most_frequent != most_common_words(wordFreqs, 3):
    raise Exception("Failed Test 1")

# Test 2
if top_five_most_frequent != most_common_words(wordFreqs, 5):
    raise Exception("Failed Test 2")

# Test 3
if top_ten_most_frequent != most_common_words(wordFreqs, 10):
    raise Exception("Failed Test 3")

print("Passes all tests!")
```

```python
#Word Lengths

word_of_length = {}

for word in words: 
    word_length = len(word)
    if word_length in word_of_length:
        word_of_length[word_length].append(word)
    else:
        word_of_length[word_length] = [word]

print(word_of_length)
```
